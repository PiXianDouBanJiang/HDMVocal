```python
import matplotlib.pyplot as plt

matplotlib.rcParams['font.size'] = 24  # 根据需要调整字号
plt.rcParams['axes.linewidth'] = 2

# 已知数据
x = ['0.01, 10', '1, 9', '1, 8', '10, 8']
y1 = [4.42, 2.36, 3.23, 6.59]
y2 = [99.80, 99.37, 98.17, 98.87]

# 创建一个图形
fig, ax1 = plt.subplots(figsize=(8, 6))

# 在左侧添加第一个 y 轴
ax1.plot(x, y1, marker='o', linestyle='-', color='#EC407A', label='Dissimilarity_Kmeans (%)')
ax1.set_xlabel('Carbidopa Dose (mg), Female Mouse No.')
# ax1.set_ylabel( color='b')  # 设置左侧 y 轴标签颜色
ax1.tick_params(axis='y', labelcolor='#EC407A')  # 设置左侧 y 轴刻度标签颜色
ax1.set_yticks([3, 4, 5, 6, 7])

# 在右侧添加第二个 y 轴
ax2 = ax1.twinx()
ax2.plot(x, y2, marker='x', linestyle='--', color='#203461', label='Classification Accuracy (%)')
# ax2.set_ylabel( color='r')  # 设置右侧 y 轴标签颜色
ax2.tick_params(axis='y', labelcolor='#203461')  # 设置右侧 y 轴刻度标签颜色
ax2.set_yticks([98.5, 99, 99.5, 100])

# 添加图例
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right', fontsize=20)

# 设置标题
# plt.title('Dual Y-axis Curve Plot')

# 自动调整子图参数，以便图像完全适合
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_double_bar_chart(x, y1, y2):
    # 设置柱状图的横坐标位置
    x_pos = np.arange(len(x))
    
    # 设置每个柱状图的宽度
    width = 0.3
    
    matplotlib.rcParams['font.size'] = 20  # 根据需要调整字号
    plt.rcParams['axes.linewidth'] = 2
    
    # 绘制两组柱状图
    plt.bar(x_pos-0.025, y1, width, label='Original', color='#26A69A', edgecolor='black')
    plt.bar(x_pos + width+0.025, y2, width, label='Screened', color='#2F7FC1', edgecolor='black')
    
    # 设置横坐标刻度和标签
    plt.xticks(x_pos + width/2, x)
    plt.yticks([0.8, 0.85, 0.9, 0.95, 1])
    
    # 设置纵坐标范围
    plt.ylim([0.8, 1])

    # # 设置坐标轴标签
    # plt.xlabel('X')
    # plt.ylabel('Y')
    
    # 添加图例
    plt.legend(fontsize=16)
    
    # 显示柱状图
    plt.show()

# 输入示例数据
x = ['Accuracy', 'Precision', 'Recall']
y1 = [0.9195176668536175,	0.9290890269151139,	0.922879177377892]
y2 = [0.9277777777777778,	0.9305555555555556,	0.9481132075471698]

# 生成并显示直方图
generate_double_bar_chart(x, y1, y2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_double_bar_chart(x, y1, y2):

    plt.figure(figsize=(6.9, 4.8))

    matplotlib.rcParams['font.size'] = 20  # 根据需要调整字号

    # 设置柱状图的横坐标位置
    x_pos = np.arange(len(x))
    
    # 设置每个柱状图的宽度
    width = 0.17
    
    # 绘制两组柱状图
    plt.bar(x_pos - width - 0.03, y1, width, color = '#2d755e', label='CatBoost', edgecolor='black')
    plt.bar(x_pos, y2, width, color = '#b6d3ff', label='XGBoost', edgecolor='black')
    plt.bar(x_pos + width +0.03, y3, width, color = '#ffe8cf', label='AdaBoost', edgecolor='black')
    plt.bar(x_pos + width * 2 +0.06, y4, width, color = '#4e274a', label='RF', edgecolor='black')
    
    # 设置横坐标刻度和标签
    plt.xticks(x_pos + width/2, x)

    # 设置纵坐标范围
    plt.ylim([0.6, 1])

    # # 设置坐标轴标签
    # plt.xlabel('M')
    # plt.ylabel('Y')
    
    # 添加图例
    plt.legend(frameon=False, loc='upper left', fontsize=16, 
    bbox_to_anchor=(1, 1)
    )

    # 显示柱状图
    plt.show()

# 输入示例数据
x = ['Accuracy', 'Precision', 'Recall']
y1 = [0.9195176668536175, 0.9290890269151139, 0.922879177377892]
y2 = [0.8853056646102075, 0.8904884318766066, 0.8983402489626556]
y3 = [0.8564217610768368, 0.8704370179948586, 0.8668714797747056]
y4 = [0.8143578238923164, 0.8390745501285347, 0.8238263503281171]

# 生成并显示直方图
generate_double_bar_chart(x, y1, y2)
```
